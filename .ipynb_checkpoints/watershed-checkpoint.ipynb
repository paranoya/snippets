# Watershed

```python
sorted_by_data = np.argsort(data.flatten())[::-1]
sorted_index = np.unravel_index(sorted_by_data, data.shape)
pix_ID = np.reshape(np.arange(data.size), data.shape)
```

```python
parent = -np.ones(data.size, dtype=int)
top = -np.ones(data.size, dtype=int)
area = np.zeros(data.size, dtype=int)

#for row, col in np.transpose(sorted_index):
for pixel in sorted_by_data:
    row, col = np.unravel_index(pixel, data.shape)
    row0 = max(row-1, 0)
    row1 = min(row+2, data.shape[0])
    col0 = max(col-1, 0)
    col1 = min(col+2, data.shape[1])
    neighbour_parents = []
    for r in range(row0, row1):
        for c in range(col0, col1):
            p = parent[np.ravel_multi_index((r, c), data.shape)]
            if p > -1:
                neighbour_parents.append(top[p])
#    if row==0:
#        print(col, neighbour_parents)

    neighbour_parents = np.unique(neighbour_parents)
    n_parents = neighbour_parents.size
    if n_parents == 0:
        selected_parent = pixel
    elif n_parents == 1:
        selected_parent = neighbour_parents[0]
    else:
        selected_parent = neighbour_parents[np.argmax(area[neighbour_parents])]
        for p in neighbour_parents:
            top[p] = top[selected_parent]
#        print(selected_parent)

    parent[pixel] = top[pixel] = selected_parent
    area[selected_parent] += 1

area = area[parent]
```

```python
im = ax1[1].imshow(parent.reshape(data.shape),
                  interpolation='nearest', origin='lower', cmap='nipy_spectral',
                 )
cb1 = fig1.colorbar(im, cax=cb1.ax)
im = ax1[2].imshow(np.log10(area).reshape(data.shape),
                  interpolation='nearest', origin='lower', cmap='nipy_spectral',
                  #vmin=0, vmax=100,
                 )
cb2 = fig1.colorbar(im, cax=cb2.ax)
```

```python
fig, ax = plt.subplots()
ax.hist(area, bins=np.arange(0, 100))
ax.set_yscale('log')
```
